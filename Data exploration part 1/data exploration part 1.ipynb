## Data Analysis Possibilities

By analyzing this employee attrition dataset, organizations can gain valuable insights into their workforce. Here are some key areas that can be analyzed:

1. **Turnover Rate:**
   - Measurement of the rate at which employees are leaving the company.
   - Calculated as a percentage of the total number of employees.

2. **Time to Attrition:**
   - Assessment of the amount of time an employee remains with the company before resigning.

3. **Demographic Analysis:**
   - Examination of attrition rates based on various demographic categories, including:
     - Gender
     - Age
     - Education level
     - Years of experience

4. **Department or Role-Based Analysis:**
   - Evaluation of attrition rates in different sections or job roles within the organization.
   - Identification of areas with higher turnover rates.

5. **Correlation Analysis:**
   - Exploration of correlations between employee attrition and other factors such as:
     - Job satisfaction
     - Work environment
     - Compensation

6. **Cost Analysis:**
   - Calculation of the costs associated with attrition, including:
     - Recruitment and training expenses for replacing departing employees.

Analyzing these aspects can help organizations identify opportunities for improvement, develop strategies to reduce staff turnover, enhance employee engagement, and increase overall employee satisfaction.

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
# url = "https://www.kaggle.com/datasets/pavansubhasht/ibm-hr-analytics-attrition-dataset"
# df = pd.read_csv(url)

df = pd.read_csv('employee attrition.csv')

# Check for missing values in the selected variables
selected_variables = ['Age', 'MonthlyIncome', 'JobSatisfaction', 'HourlyRate', 'PerformanceRating', 'NumCompaniesWorked', 'DistanceFromHome', 'YearsInCurrentRole']

# Check for missing values
missing_values = df[selected_variables].isnull().sum()

# 3. Perform univariate descriptive statistics on selected variables
summary_statistics = df[selected_variables].describe()

# Visualize the data using histograms and boxplots
plt.figure(figsize=(14, 10))
for i, variable in enumerate(selected_variables, 1):
    plt.subplot(4, 2, i)
    sns.histplot(df[variable], bins=20, kde=True)
    plt.title(f'{variable} Distribution')

plt.tight_layout()
plt.show()

# 4. Check for outliers using boxplots
plt.figure(figsize=(14, 10))
for i, variable in enumerate(selected_variables, 1):
    plt.subplot(4, 2, i)
    sns.boxplot(x=df[variable])
    plt.title(f'{variable} Boxplot')

plt.tight_layout()
plt.show()



# Updated research questions
research_questions = [
    "Is there a relationship between JobSatisfaction and Attrition?",
    "Do employees with higher MonthlyIncome have lower Attrition rates?",
    "Does DistanceFromHome impact Attrition rates?",
    "Is there a significant difference in JobSatisfaction between different JobRoles?",
]
```
